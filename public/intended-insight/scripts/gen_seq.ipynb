```python
import revisitpy as rvt
import itertools
import json
```

```python
# Trying .permute

seq = rvt.sequence(order="fixed").permute(
    factors=[{'chart': 'cdc'}, {'chart': 'who'}, {'chart': 'covidnet'}, {'chart': 'wapo'}, {'chart': 'map'}],
    order='latinSquare'
).permute(
    factors=[{'label': 'none'}, {'label': 'warning'}, {'label': 'counter'}, {'label': 'guide'}, {'label': 'all'}],
    order='latinSquare',
    numSamples=1
)

print(seq)
```

```python
# Brute the sequence

charts = ["cdc", "who", "covidnet", "map", "wapo"]
labels = ["none", "warning", "counter", "guide", "all"]

# all permutations
valid_sets = list(itertools.permutations(labels))

def format_pair(chart, label):
    base_id = f"{chart}-{label}"
    return {
        "order": "fixed",
        "components": [
            f"{base_id}-insight",
            f"{base_id}-trust",
            f"{base_id}-agree"
        ]
    }

formatted_sets = []
for perm in valid_sets:
    set_dict = [format_pair(chart, label) for chart, label in zip(charts, perm)]
    formatted_sets.append(set_dict)

for i, s in enumerate(formatted_sets, start=1):
    print(json.dumps({"order": "random", "components": s}, indent=2))
    print(',')
```

```python
# Brute the component definitions too

config = {}

for chart, label in itertools.product(charts, labels):
    for component in ["insight", "trust", "agree"]:
        key = f"{chart}-{label}-{component}"
        config[key] = {
            "baseComponent": f"chart-stimulus-{component}",
            "parameters": {
                "exampleId": chart,
                "textType": label
            }
        }

print(json.dumps(config, indent=2))
```
